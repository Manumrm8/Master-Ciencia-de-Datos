```python
import numpy as np
from sklearn.datasets import load_iris, load_breast_cancer
from sklearn import tree
from os import system
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score
from sklearn import tree

```

# Con iris

```python
iris = load_iris()  # load iris data

n_classes  = iris.target_names.size

X = iris.data
y = iris.target
split=np.random.get_state()
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3)

ynames = iris.target_names
fnames = iris.feature_names
```

```python
tamaños_de_arbol=14
p_train=np.empty((tamaños_de_arbol, 2))
p_test=np.empty((tamaños_de_arbol, 2))


str=np.random.get_state()
for i in range(tamaños_de_arbol):
    np.random.set_state(str) 
    clf = DecisionTreeClassifier( 
    criterion = 'entropy', #  default 'gini'
    splitter  = 'best' , #  default 'best'
    max_depth = i+1,      )

    clf.fit(X_train, y_train)
    entrenamiento=clf.predict(X_train)
    testeo=clf.predict(X_test)
    p_train[i,1]=accuracy_score(y_train, entrenamiento)
    p_test[i,1]=accuracy_score(y_test, testeo)
    a=clf.tree_.node_count
    p_train[i,0]=a
    p_test[i,0]=a
    

plt.plot(p_train[:,0],p_train[:,1], label="Train",marker='o', linestyle='-')
plt.plot(p_test[:,0],p_test[:,1], label="Test",marker='o', linestyle='-')
plt.xlabel('Número de Nodos')
plt.ylabel('Precisión')
plt.title("Iris criterio entropía")
plt.legend()


```

```python
plt.figure(figsize=(12,8))

tree.plot_tree(clf.fit(X_train, y_train)) 
```

# Poda y características con iris

```python
X = iris.data
y = iris.target
np.random.set_state(split) 
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3)
```

```python
for i in range(tamaños_de_arbol):
    np.random.set_state(str) 
    clf = DecisionTreeClassifier( 
    criterion = 'entropy', #  default 'gini'
    splitter  = 'best' , #  default 'best'
    max_depth = i+1,      )

    clf.fit(X_train, y_train)
    entrenamiento=clf.predict(X_train)
    testeo=clf.predict(X_test)
    p_train[i,1]=accuracy_score(y_train, entrenamiento)
    p_test[i,1]=accuracy_score(y_test, testeo)
    a=clf.tree_.node_count
    p_train[i,0]=a
    p_test[i,0]=a
    
```

```python
flag = False
prof = 2
numero_hojas_p = []
resultados_poda = []

while(not(flag)):
    np.random.set_state(split) 
    clf= DecisionTreeClassifier( 
    criterion = 'entropy', #  default 'gini'
    splitter  = 'best' , #  default 'best'
    max_depth = i+1,ccp_alpha = 0.01      )

    clf.fit(X_train, y_train)

    numero_hojas_p.append(clf.get_n_leaves())

    score_test = clf.score(X_test,y_test)
    resultados_poda.append(score_test)
    prof+=1

    if(prof>15):
        flag=True

# Graficar utilizando numero_hojas como el eje x para ambas gráficas
plt.plot(p_train[:,0],p_train[:,1], label="Train",marker='o', linestyle='-')
plt.plot(p_test[:,0],p_test[:,1], label="Test",marker='o', linestyle='-')
plt.plot(numero_hojas_p, resultados_poda, label='Test', marker="o",linestyle="-",color="green")

# Agregar etiquetas y leyenda
plt.xlabel('Número de Hojas')
plt.ylabel('Score')
plt.title('Valores del score del train y test por numero de hojas')
plt.legend()

# Mostrar la gráfica
plt.show()
```
